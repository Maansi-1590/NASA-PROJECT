```python
asteroids_data = []
target = 10000
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2025-05-01&end_date=2025-05-08&api_key={API_KEY}"
while len(asteroids_data) < target:
  response = requests.get(url)
  data = response.json()
  details = data['near_earth_objects']
  for date, asteroids in details.items():
      for ast in asteroids:
          asteroids_data.append(dict(
              id = int(ast['id']),
              name = ast['name'],
              neo_reference_id = ast['neo_reference_id'],
              absolute_magnitude_h = ast['absolute_magnitude_h'],
              estimated_diameter_min_km = ast['estimated_diameter']['kilometers']['estimated_diameter_min'],
              estimated_diameter_max_km = ast['estimated_diameter']['kilometers']['estimated_diameter_max'],
              is_potentially_hazardous_asteroid = ast['is_potentially_hazardous_asteroid'],
              close_approach_data = ast['close_approach_data'],
              relative_velocity_kmph = ast['close_approach_data'][0]['relative_velocity']['kilometers_per_hour'],
              astronomical = ast['close_approach_data'][0]['miss_distance']['astronomical'],
              miss_distance_km = ast['close_approach_data'][0]['miss_distance']['kilometers'],
              miss_distance_lunar = ast['close_approach_data'][0]['miss_distance']['lunar'],
              orbiting_body = ast['close_approach_data'][0]['orbiting_body']
          ))

          if len(asteroids_data) >= target:
              break
      if len(asteroids_data) >= target:
              break
  url = data['links']['next']
```

```python
from datetime import datetime
import sqlite3

connection = sqlite3.connect("nasa.db")
cursor = connection.cursor()


insert_query = "INSERT INTO close_approach VALUES (?, ?, ?, ?, ?, ?, ?)"

for asteroid in asteroids_data:
    neo_id = asteroid['neo_reference_id']

    for i in asteroid['close_approach_data']:
        try:
            formatted_date = datetime.strptime(i['close_approach_date'], "%Y-%m-%d").date()
            velocity = float(i['relative_velocity']['kilometers_per_hour'])
            astro = float(i['miss_distance']['astronomical'])
            miss_dist = float(i['miss_distance']['kilometers'])
            lunar = float(i['miss_distance']['lunar'])
            body = i['orbiting_body']

            values = (neo_id, formatted_date, velocity, astro, miss_dist, lunar, body)

            cursor.execute(insert_query, values)

        except Exception as e:
            print(f"Error processing asteroid data: {e}")
            continue # Skip to the next close approach data if an error occurs


connection.commit()
```

```python
cursor.execute('''
    SELECT
        a.name,
        MIN(ca.close_approach_date) AS first_approach,
        MAX(ca.close_approach_date) AS last_approach,
        JULIANDAY(MAX(ca.close_approach_date)) - JULIANDAY(MIN(ca.close_approach_date)) AS days_between
    FROM close_approach ca
    JOIN asteroids a ON ca.neo_refrence_id = a.id
    GROUP BY a.name
    ORDER BY days_between DESC
    LIMIT 10;
''')

longest_span_asteroids = pd.DataFrame(cursor.fetchall(), columns=[i[0] for i in cursor.description])
print(longest_span_asteroids)
```

```python
%%writefile nasa_asteroids.py
import streamlit as st
import pandas as pd
import sqlite3
from streamlit_option_menu import option_menu

# Title
st.title("NASA ASTEROIDS DATA " )

# Connect to the database
connection = sqlite3.connect('nasa.db')
cursor = connection.cursor()

# Sidebar menu
with st.sidebar:
    selected = option_menu('Main Menu', ['Home', 'Queries','Custom Filter'],
                           icons=['house', 'gear','filter-circle'],
                           menu_icon='cast', default_index=0)

# Home section
if selected == 'Home':
    st.write("Welcome to the NASA Asteroids Explorer")

# Queries section
elif selected == 'Queries':
    query_option = st.selectbox("Choose a Query", [
        '1. Count approaches per year',
        '2. List asteroids with multiple approaches in a year',
        '3. Asteroids with only one recorded close approach',
        '4. Asteroids with longest time span between approaches',
        '5. Asteroids with smallest average miss distance',
        '6. Count how many times each asteroid has approached Earth',
        '7. Average velocity of each asteroid over multiple approaches',
        '8. Top 10 fastest asteroids',
        '9. Potentially hazardous asteroids with >3 approaches',
        '10. Month with the most asteroid approaches',
        '11. Asteroid with the fastest ever approach speed',
        '12. Sort asteroids by maximum estimated diameter (descending)',
        '13. Asteroid whose closest approach is getting nearer over time',
        '14. Name each asteroid with date and miss distance of closest approach',
        '15. Asteroids with velocity > 50,000 km/h',
        '16. Count how many approaches happened per month',
        '17. Asteroid with highest brightness (lowest magnitude)',
        '18. Number of hazardous vs non-hazardous asteroids',
        '19. Asteroids that passed closer than the Moon (less than 1 LD)',
        '20. Asteroids that came within 0.05 AU (astronomical distance)'
    ])


    # Handle each query
    if query_option == '1. Count approaches per year':
        cursor.execute('''
            SELECT
                strftime('%Y', close_approach_date) AS year,
                COUNT(*) AS approach_count
            FROM close_approach
            GROUP BY year
            ORDER BY approach_count DESC;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Year', 'Approach Count'])
        st.dataframe(df)

    elif query_option == '2. List asteroids with multiple approaches in a year':
        cursor.execute('''
            SELECT
                a.name,
                strftime('%Y', ca.close_approach_date) AS year,
                COUNT(*) AS approach_count
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name, year
            HAVING approach_count > 1
            ORDER BY approach_count DESC;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Year', 'Approach Count'])
        st.dataframe(df)

    elif query_option == '3. Asteroids with only one recorded close approach':
        cursor.execute('''
            SELECT a.name, COUNT(*) AS approach_count
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name
            HAVING approach_count = 1;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Approach Count'])
        st.dataframe(df)

    elif query_option == '4. Asteroids with longest time span between approaches':
        cursor.execute('''
            SELECT
                a.name,
                MIN(ca.close_approach_date) AS first_approach,
                MAX(ca.close_approach_date) AS last_approach,
                JULIANDAY(MAX(ca.close_approach_date)) - JULIANDAY(MIN(ca.close_approach_date)) AS span_days
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name
            ORDER BY span_days DESC
            LIMIT 10;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'First Approach', 'Last Approach', 'Time Span (days)'])
        st.dataframe(df)

    elif query_option == '5. Asteroids with smallest average miss distance':
        cursor.execute('''
            SELECT
                a.name,
                AVG(ca.miss_distance_km) AS avg_miss_distance
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name
            ORDER BY avg_miss_distance ASC
            LIMIT 10;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Average Miss Distance (km)'])
        st.dataframe(df)

    elif query_option == '6. Count how many times each asteroid has approached Earth':
        cursor.execute('''
            SELECT a.name, COUNT(*) AS approach_count
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name
            ORDER BY approach_count DESC;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Approach Count'])
        st.dataframe(df)

    elif query_option == '7. Average velocity of each asteroid over multiple approaches':
        cursor.execute('''
            SELECT a.name, AVG(ca.relative_velocity_kmph) AS avg_velocity
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name
            ORDER BY avg_velocity DESC;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Average Velocity (km/h)'])
        st.dataframe(df)

    elif query_option == '8. Top 10 fastest asteroids':
        cursor.execute('''
            SELECT a.name, MAX(ca.relative_velocity_kmph) AS max_velocity
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            GROUP BY a.name
            ORDER BY max_velocity DESC
            LIMIT 10;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Max Velocity (km/h)'])
        st.dataframe(df)

    elif query_option == '9. Potentially hazardous asteroids with >3 approaches':
        cursor.execute('''
            SELECT a.name, COUNT(*) AS approach_count
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            WHERE a.is_potentially_hazardous_asteroid = 1
            GROUP BY a.name
            HAVING approach_count > 3
            ORDER BY approach_count DESC;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Approach Count'])
        st.dataframe(df)

    elif query_option == '10. Month with the most asteroid approaches':
        cursor.execute('''
            SELECT strftime('%m', close_approach_date) AS month, COUNT(*) AS approach_count
            FROM close_approach
            GROUP BY month
            ORDER BY approach_count DESC
            LIMIT 1;
        ''')
        result = cursor.fetchone()
        st.write(f"Month with the most approaches: {result[0]} ({result[1]} approaches)")

    elif query_option == '11. Asteroid with the fastest ever approach speed':
        cursor.execute('''
            SELECT a.name, ca.relative_velocity_kmph, ca.close_approach_date
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            ORDER BY ca.relative_velocity_kmph DESC
            LIMIT 1;
        ''')
        result = cursor.fetchone()
        st.write(f"Asteroid: {result[0]}, Speed: {result[1]} km/h, Date: {result[2]}")

    elif query_option == '12. Sort asteroids by maximum estimated diameter (descending)':
        cursor.execute('''
            SELECT name, MAX(estimated_diameter_max_km) AS max_diameter
            FROM asteroids
            GROUP BY name
            ORDER BY max_diameter DESC;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Max Diameter (km)'])
        st.dataframe(df)

    elif query_option == '13. Asteroid whose closest approach is getting nearer over time':
        cursor.execute('''
            SELECT a.name,
                (SELECT ca1.miss_distance_km FROM close_approach ca1
                 WHERE ca1.neo_refrence_id = a.id
                 ORDER BY ca1.close_approach_date ASC LIMIT 1) AS first_distance,
                (SELECT ca2.miss_distance_km FROM close_approach ca2
                 WHERE ca2.neo_refrence_id = a.id
                 ORDER BY ca2.close_approach_date DESC LIMIT 1) AS last_distance
            FROM asteroids a
            GROUP BY a.name
            HAVING last_distance < first_distance;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'First Miss Distance (km)', 'Last Miss Distance (km)'])
        st.dataframe(df)

    elif query_option == '14. Name each asteroid with date and miss distance of closest approach':
        cursor.execute('''
            SELECT a.name, ca.close_approach_date, ca.miss_distance_km
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            WHERE (a.id, ca.miss_distance_km) IN (
                SELECT neo_reference_id, MIN(miss_distance_km)
                FROM close_approach
                GROUP BY neo_refrence_id
            );
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Closest Approach Date', 'Miss Distance (km)'])
        st.dataframe(df)

    elif query_option == '15. Asteroids with velocity > 50,000 km/h':
        cursor.execute('''
            SELECT DISTINCT a.name
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            WHERE ca.relative_velocity_kmph > 50000;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name'])
        st.dataframe(df)

    elif query_option == '16. Count how many approaches happened per month':
        cursor.execute('''
            SELECT strftime('%Y-%m', close_approach_date) AS month, COUNT(*) AS approach_count
            FROM close_approach
            GROUP BY month
            ORDER BY month;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Month', 'Approach Count'])
        st.dataframe(df)

    elif query_option == '17. Asteroid with highest brightness (lowest magnitude)':
        cursor.execute('''
            SELECT a.name, MIN(a.absolute_magnitude_h)
            FROM asteroids a
            GROUP BY a.name
            ORDER BY MIN(a.absolute_magnitude_h) ASC
            LIMIT 1;
        ''')
        result = cursor.fetchone()
        st.write(f"Brightest Asteroid: {result[0]} with magnitude {result[1]}")

    elif query_option == '18. Number of hazardous vs non-hazardous asteroids':
        cursor.execute('''
            SELECT is_potentially_hazardous_asteroid, COUNT(*)
            FROM asteroids
            GROUP BY is_potentially_hazardous_asteroid;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Is Hazardous', 'Asteroid Count'])
        df['Is Hazardous'] = df['Is Hazardous'].map({0: 'No', 1: 'Yes'})
        st.dataframe(df)

    elif query_option == '19. Asteroids that passed closer than the Moon (less than 1 LD)':
        cursor.execute('''
            SELECT a.name, ca.close_approach_date, ca.miss_distance_lunar
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            WHERE ca.miss_distance_lunar < 1;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Date', 'Miss Distance (LD)'])
        st.dataframe(df)

    elif query_option == '20. Asteroids that came within 0.05 AU (astronomical distance)':
        cursor.execute('''
            SELECT a.name, ca.close_approach_date, ca.astronomical
            FROM close_approach ca
            JOIN asteroids a ON ca.neo_refrence_id = a.id
            WHERE ca.astronomical < 0.05;
        ''')
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=['Asteroid Name', 'Date', 'Miss Distance (AU)'])
        st.dataframe(df)
elif selected == "Custom Filter":
    st.header(" Filter Asteroids Based on Your Criteria")

    # Input Widgets
    date = st.date_input("Close Approach Date", value=None)
    au_range = st.slider("Miss Distance (AU)", 0.0, 1.5, (0.0, 1.0), 0.01)
    lunar_range = st.slider("Lunar Distance", 0.0, 400.0, (0.0, 50.0), 1.0)
    velocity_range = st.slider("Relative Velocity (km/hr)", 0.0, 150000.0, (0.0, 100000.0), 500.0)
    diameter_range = st.slider("Estimated Diameter (km)", 0.0, 10.0, (0.0, 2.0), 0.1)
    hazard = st.selectbox("Hazardous?", ["All", "Yes", "No"])

    # Build dynamic SQL query
    query = """
        SELECT a.name, ca.close_approach_date, ca.astronomical, ca.miss_distance_lunar,
               ca.relative_velocity_kmph, a.estimated_diameter_min_km, a.estimated_diameter_max_km,
               a.is_potentially_hazardous_asteroid
        FROM close_approach ca
        JOIN asteroids a ON ca.neo_reference_id = a.id
        WHERE 1 = 1
    """

    if date:
        query += f" AND ca.close_approach_date = '{date}'"
    query += f" AND ca.astronomical BETWEEN {au_range[0]} AND {au_range[1]}"
    query += f" AND ca.miss_distance_lunar BETWEEN {lunar_range[0]} AND {lunar_range[1]}"
    query += f" AND ca.relative_velocity_kmph BETWEEN {velocity_range[0]} AND {velocity_range[1]}"
    query += f" AND a.estimated_diameter_min_km >= {diameter_range[0]}"
    query += f" AND a.estimated_diameter_max_km <= {diameter_range[1]}"
    if hazard == "Yes":
        query += " AND a.is_potentially_hazardous_asteroid = 1"
    elif hazard == "No":
        query += " AND a.is_potentially_hazardous_asteroid = 0"


    df_filtered = pd.read_sql_query(query, connection)
    st.subheader(" Filtered Results")
    st.dataframe(df_filtered)


connection.close()
```
